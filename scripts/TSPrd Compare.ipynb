```python
from tsprd_data import *
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format 
```

### Resultados da nova execução

```python
df_new = aggregate_data(read_execution_data("output_oldinter")).query("set == 'Solomon' and n <= 20")
df_new.loc[df_new["sol_time"] < 1, "sol_time"] = 1
df_new.describe()
```


### Resultados atuais

```python
df_old = aggregate_data(read_execution_data("output_ref"))
df_old = df_old.loc[df_new.index]
df_old.loc[df_old["sol_time"] < 1, "sol_time"] = 1
df_old.describe()
```

### Quantidade de melhores, iguais e piores e gaps em cada coluna

```python
columns = ["best_obj","avg_obj", "exec_time", "sol_time"]
new, old = df_new[columns], df_old[columns]
old.loc[old["sol_time"] == 0, "sol_time"] = 1

agg = pd.concat({"better": new < old, "equal": new == old, "worse": new > old}, axis=1).sum().unstack(level=0)
agg["gap"] = ((100 * new / old) - 100).mean()
agg
```

```python
df_old[["best_obj","avg_obj"]].compare(df_new[["best_obj","avg_obj"]])
```
